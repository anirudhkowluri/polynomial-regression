```python
#train test split#
```

```python
X = df[['LSTAT' ,'RM']]
y = df['MEDV']
X.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=5)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
lr  = LinearRegression()
lr.fit(x_train, y_train)
```

```python
# prediction on train data
```

```python
lr_predict_train = lr.predict(x_train)
rmse = np.sqrt(mean_squared_error(y_train, lr_predict_train))
print("RMSE: ", rmse)
r2 = r2_score(y_train, lr_predict_train)
print("R2: ", r2)
```

```python
#prediction on test data
```

```python
lr_predict_test = lr.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, lr_predict_test))
print(rmse)
r2 = r2_score(y_test, lr_predict_test)
print(r2)
```

```python
# to find the ideal best fit line we visualize the data#
```

```python
plt.scatter(y_test, lr_predict_test)
plt.plot([lr_predict_test.min(), lr_predict_test.max(), ], [lr_predict_test.min(), lr_predict_test.max()], color='red') 
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted")
plt.show()
```
